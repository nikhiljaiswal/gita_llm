````python
import pandas as pd

dff=pd.DataFrame()
for f in fl:
    if f.endswith('xlsx') == False:
        continue
    f=os.path.join(dir,f)
    print(f)
    df = pd.read_excel(f,sheet_name='WM-FL', engine='openpyxl')
    #df = pd.read_excel('data/TestCases/ENG_TC_WM_FL_FAQ_V1.3.xlsx',sheet_name='WM-FL')
    df = df[df['Part no']=='MFL71521482']
    df = df.sample(frac=1, random_state=1).reset_index(drop=True)
    df = df.head(10)
    dff = pd.concat([dff,df])
dff.head(10)
````
